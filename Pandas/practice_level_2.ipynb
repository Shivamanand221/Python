```python
#Question 1: Filtering Rows

df = pd.DataFrame({
    "Name": ["Aman", "Riya", "Karan", "Priya"],
    "Age": [21, 22, 20, 24],
    "City": ["Delhi", "Mumbai", "Delhi", "Pune"]
})
print(df["Age"]>21)
print(df[df["City"]=="Delhi"])
print(df[(df["City"]=="Delhi") & (df["Age"]>20)])
```
