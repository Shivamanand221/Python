```python
#Question 1: Filtering Rows

df = pd.DataFrame({
    "Name": ["Aman", "Riya", "Karan", "Priya"],
    "Age": [21, 22, 20, 24],
    "City": ["Delhi", "Mumbai", "Delhi", "Pune"]
})
print(df["Age"]>21)
print(df[df["City"]=="Delhi"])
print(df[(df["City"]=="Delhi") & (df["Age"]>20)])
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name": ["Aman", "Riya", "Karan", "Priya"],
    "Marks": [85, np.nan, 78, 92]
})
```

```python
# 2. Count missing values
print("Missing Values:")
print(df.isnull().sum())
```

```python
# 3. Replace missing values
df = pd.DataFrame({
    "Name": ["Aman", "Riya", "Karan"],
    "Age": [21, 22, 20],
    "City": ["Delhi", "Mumbai", "Pune"]
})
df["Marks"] = df["Marks"].fillna(0)

print("\nDataFrame after filling missing values:")
print(df)
```

```python
# 4. Average marks calculation
average_marks = df["Marks"].mean()

print("\nAverage Marks:")
print(average_marks)
```
