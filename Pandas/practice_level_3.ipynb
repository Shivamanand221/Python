```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name": ["Aman", "Riya", "Karan", "Priya", "Rahul"],
    "Age": [21, 22, 20, 24, 23],
    "Marks": [85, 90, 78, 92, 65]
})
```

```python
#Question 1: Multiple Filters
print(df[df["Marks"]>=80])
print(df[(df["Age"]>21) & (df["Marks"]>80)])
print(df.loc[df["Marks"]>80, ["Name", "Marks"]])
```

```python
#Question 2: Sorting Data
df = pd.DataFrame({
    "Name": ["Aman", "Riya", "Karan", "Priya"],
    "Marks": [85, 90, 78, 92],
    "Age": [21, 22, 20, 24]
})
print(df.sort_values(by="Marks"))
print(df.sort_values(by="Marks", ascending=False))
print(df.sort_values(by=["Marks", "Age"], ascending=[False, True]))
```

```python
#Question 3: Missing Values

df = pd.DataFrame({
    "Name": ["Aman", "Riya", None, "Priya"],
    "Marks": [85, np.nan, 78, 92]
})

print(df.isnull().sum())
print("Total missing:", df.isnull().sum().sum())
print()
```
