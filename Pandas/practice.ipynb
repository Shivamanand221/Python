```python
df3 = pd.DataFrame({'ID': [1,2,3], 'Name':['A', 'B', 'C']})
df4 = pd.DataFrame({'ID': [1,2,2,4], 'Score': [88, 96, 77, 79]})
print(df3)
print(df4)
```

```python
pd.concat([df3, df4], axis=0)
```

```python
pd.concat([df3, df4], axis=1)
```

```python
print(pd.merge(df3, df4, how='inner', on='ID'))
print(pd.merge(df3, df4, how='right', on='ID'))
print(pd.merge(df3, df4, how='left', on='ID'))
```

```python
print(pd.merge(df3, df4, how='inner', left_on='ID', right_on='ID'))
```
