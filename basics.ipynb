```python
def countdown(n):
    if n==0:
        return
    print(n)
    countdown(n-1)

countdown(5)
```

```python
def factorial(n):
    if n==1:
        return 1
    
    return n*factorial(n-1)

factorial(5)
    
```

```python
def sum(n):
    if n==0:
        return 0
    
    return n+sum(n-1)

sum(10)
```
