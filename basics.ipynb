```python
class animal:
    def eat(self):
        print("eating")
    
    def sleep(self):
        print("sleeping")
    
class dog(animal):
    def bark(self):
        print("barking")

dog1=dog()
dog1.bark()
dog1.eat()
dog1.sleep()

class newdog(dog):
    def run(self):
        print("run")

dog2=newdog()
dog2.run()
dog2.bark()
dog2.eat()
dog2.sleep()
```

```python
#Method Overriding

class animal:
    def sound(self):
        print("Animal sound")

class dog(animal):
    def sound(self):
        print("Bark")

dog1=dog()
dog1.sound()

dog2=animal()
dog2.sound()
```

```python
class Animal:

    def sound(self):
        print("Animal Sound")


class Dog(Animal):

    def sound(self):
        super().sound()
        print("Bark")

dog = Dog()
dog.sound()
```

```python
class animal():

    def __init__(self):
        print("animal created")

class dog(animal):
    def __init__(self):
        super().__init__()
        print("dog created")

dog1=dog()
```

```python
class animal:
    def __init__(self, name):
        self.name=name

class dog(animal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed= breed

dog=dog("tom", "labrador")
print(dog.name)
print(dog.breed)
```

```python
#Multiple Inheritance
class camara:
    def click(self):
        print("photo")

class phone:
    def call(self):
        print("calling")

class smartphone(camara, phone):
    pass

phone1= smartphone()
phone1.call()
phone1.click()
```
